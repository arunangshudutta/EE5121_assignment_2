```python
import cvxpy as cp
import numpy as np
```

# Graph (i)

```python
G_edges = [(1,2),(1,3),(1,4),(2,3),(2,4),(3,4)]
n = 4

E = [(u-1, v-1) for (u, v) in G_edges]
```

```python
# Primal problem solver

# Variables
G = cp.Variable((n, n), symmetric=True)
rho = cp.Variable()

constraints = []

# PSD constraint
constraints += [G >> 0]

# Diagonal = 1
for v in range(n):
    constraints += [G[v, v] == 1]

# Edge constraints G_uv <= rho
for (u, v) in E:
    constraints += [G[u, v] <= rho]

# Objective
objective = cp.Minimize(rho)

prob = cp.Problem(objective, constraints)
prob.solve(solver=cp.SCS, verbose=False)

print("==== PRIMAL ====")
print("rho* =", rho.value)
print("G* =")
print(G.value)
print("Primal function optimum value rho =", rho.value)
```

```python
# Dual problem solver
m = len(E)

# Dual variables
lam = cp.Variable(n)
alpha = cp.Variable(m, nonneg=True)    # α_uv >= 0

# Construct S
S = cp.Variable((n, n), symmetric=True)

constraints = []

# S = diag(lam) + Σ α_uv (E_uv + E_vu)
expr = cp.diag(lam)

for idx, (u, v) in enumerate(E):
    A = np.zeros((n, n))
    A[u, v] = 1
    A[v, u] = 1
    expr += alpha[idx] * A

constraints += [S == expr]

# sum α_uv = 1
constraints += [cp.sum(alpha) == 1]

# PSD constraint
constraints += [S >> 0]

# Objective: maximize  -sum(lam)
objective = cp.Maximize(-cp.sum(lam))

prob = cp.Problem(objective, constraints)
prob.solve(solver=cp.SCS, verbose=False)

print("==== DUAL ====")
print("Dual optimum function value =")
print(-np.sum(lam.value))
```

```python
print("Duality Gap =", abs(-np.sum(lam.value) - rho.value))
```
